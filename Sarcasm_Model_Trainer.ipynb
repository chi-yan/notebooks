```python
import json

f = open('dataset.json',)
data = []
x = [] #sentences
y = [] #labels
with open('dataset.json') as f:
    for line in f:
        #data.append(json.loads(line))
        x.append(json.loads(line)['headline'])
        y.append(json.loads(line)['is_sarcastic'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=42)
```

```python
 embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
x_train_embeddings = embed(x_train)
```

```python
#print(x_train_embeddings.shape)
print(len(y_train))
```

```python
print(x_train_embeddings.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train_embeddings, y_train)
```

```python
logreg.intercept_
```

```python
print('Accuracy of Logistic regression classifier on train set: {:.4f}'
     .format(logreg.score(x_train_embeddings, y_train)))   #(true P + true N) / (EVERYTHING)
```

```python
x_test_embeddings = embed(x_test)
```

```python
print('Accuracy of Logistic regression classifier on test set: {:.4f}'
     .format(logreg.score(x_test_embeddings, y_test)))   #(true P + true N) / (EVERYTHING)
```
